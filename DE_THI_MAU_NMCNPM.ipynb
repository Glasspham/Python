Câu 1(2đ) Viết hàm kiểm tra xem một chuỗi có phải là mật khẩu hợp lệ hay không. Mật khẩu hợp lệ phải có từ 8 - 10 ký tự, đồng thời chỉ chứa các ký tự chữ cái và chữ số.
Tiêu đề của hàm là: def checkPass(pw):


```python
#  SV thực thi mã code câu 1 ở đây, dữ nguyên output
def checkPass(pw):
    if 8 <= len(pw) <= 10:
        if all(c.isalnum() for c in pw):
            return True
        else:
            return False
    else:
        return False

password = "Abcd1234"
result = checkPass(password)

if result:
    print("Mật khẩu hợp lệ.")
else:
    print("Mật khẩu không hợp lệ.")
```
